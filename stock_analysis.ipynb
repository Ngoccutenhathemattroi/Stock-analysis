##**CLUSTERING**

In this context, I am employing the Elbow method for identifying the most suitable number of clusters in a dataset. This technique is commonly utilized in cluster analysis and entails plotting the explained variation against the number of clusters. The optimal number of clusters is then determined by identifying the "elbow" of the curve, which is the point where further increasing the number of clusters does not significantly improve the explained variation. By utilizing the Elbow method, we can determine the most appropriate number of clusters to use, which can enhance the accuracy and interpretability of the clustering outcomes.

A good rule of thumb is to choose an elbow in the inertia plot , that is , a point where the inertia begins to drop slowly. In this case,following this criterion ,clusters = 2 is a good choice for this dataset.

```python
from sklearn.cluster import KMeans

sns.set_theme()

# Assume X is your data
distortions = []
K = range(1, 5)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_day.values)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method')
plt.show()
```

```python

# Extract features for clustering
X_line = df_day['Gia KL'].values.reshape(-1,1)
X_bar = df_day[['TB_ban_kl_1', 'TB_mua_kl_1']].values

# Normalize features
X_line_norm = (X_line - np.mean(X_line)) / np.std(X_line)
X_bar_norm = (X_bar - np.mean(X_bar, axis=0)) / np.std(X_bar, axis=0)

# Concatenate normalized features
X_norm = np.concatenate([X_line_norm, X_bar_norm], axis=1)

# Perform K-means clustering
kmeans = KMeans(n_clusters=2, random_state=0).fit(X_norm)
labels = kmeans.labels_

# Plot clustered data
fig, ax1 = plt.subplots(figsize=(10, 6))

for i in range(2):
    ax1.plot(df_day.index[labels == i], df_day['Gia KL'][labels == i],
             color=f'C{i}', label=f'Cluster {i+1}')

ax1.set_ylabel('Gia KL', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
for i in range(2):
    ax2.bar(df_day.index[labels == i], df_day['TB_ban_kl_1'][labels == i],
            color=f'C{i}', alpha=0.5, label=f'Mean_ban_kl_1 - GROUP {i+1}')
    ax2.bar(df_day.index[labels == i], df_day['TB_mua_kl_1'][labels == i],
            color=f'C{i}', alpha=0.5, label=f'Mean_mua_kl_1 - GROUP{i+1}')

ax2.set_ylabel('TB_KL', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Define time axis
ax1.xaxis.set_major_locator(plt.MaxNLocator(7))
ax1.xaxis.set_tick_params(rotation=60)

# make title
plt.title('Price based on selling and purchasing volume of stock Chart')
plt.legend(loc='upper left')
plt.show()
```

<h3><b>REMOVE OUTLIERS</b><h3>

```python
def remove_outliers(df, column_name, iqr_coef=3):
    q1 = df[column_name].quantile(0.25)
    q3 = df[column_name].quantile(0.75)
    iqr = q3 - q1
    min_val = q1 - iqr_coef * iqr
    max_val = q3 + iqr_coef * iqr
    return df[(df[column_name] >= min_val) & (df[column_name] <= max_val)]

```

###**DEFINING NUMBER OF CLUSTERS USING ELBOW METHOD**

There are two metrics used to evaluate the quality of a clustering method:

+ Distortion: This is the average of the squared distances from the cluster centers of the respective clusters. Generally, the Euclidean distance metric is utilized in this calculation.

+ Inertia: This is the sum of squared distances of samples to their closest cluster center.

We employ inertia to assess the clustering method based on two criteria:

+ Measures how spread out the clusters are, with a lower score indicating better results.

+ Measures the distance from a sample to the centroid of its cluster.

```python
df_corr = df_price.corr()
```

```python
from sklearn.cluster import KMeans

sns.set_theme()

# assume X is your data
distortions = []
K = range(1, 5)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_corr.values.reshape(-1,1))
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('The Elbow Method')
plt.show()
```

###**MODEL EVALUATION**

At first, we have to choose appropriate numbers of clusters and this is one of the biggest problems with KMeans. And in this case we choose silhouette analysis, I think the official sklearn documentation does a beautiful job of summarizing it.

Silhouette analysis can be used to study the separation distance between the resulting clusters. The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters and thus provides a way to assess parameters like the number of clusters visually. This measure has a range of (-1, 1).

Silhouette coefficients (as these values are referred to as) near +1 indicate that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.

And this is my reference link:
https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

WITHOUT PCA

```python
from sklearn.cluster import KMeans
from sklearn import metrics

# define a dictionary that contains all of our relevant info.
results_dict = {}

# define how many clusters
num_of_clusters = 8

# run through each instance of K
for k in range(2, num_of_clusters):

    # define the next dictionary to hold all the results of this run.
    results_dict[k] = {}

    # create an instance of the model, and fit the training data to it.
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_train_standard)

    # define the silhouette score
    sil_score = metrics.silhouette_score(X_train_standard, kmeans.labels_, metric='euclidean')

    # store the different metrics
    results_dict[k]['silhouette_score'] = sil_score
    results_dict[k]['score'] = kmeans.score
    results_dict[k]['model'] = kmeans
```

WITHIN PCA

```python
# define a dictionary that contains all of our relevant info.
results_dict_pca = {}

# define how many clusters
num_of_clusters = 8

# run through each instance of K
for k in range(2, num_of_clusters):

    # define the next dictionary to hold all the results of this run.
    results_dict_pca[k] = {}

    # create an instance of the model, and fit the training data to it.
    kmeans = KMeans(n_clusters=k, random_state=0).fit(pca_dataset)

    # define the silhouette score
    sil_score = metrics.silhouette_score(pca_dataset, kmeans.labels_, metric='euclidean')

    # store the different metrics
    results_dict_pca[k]['silhouette_score'] = sil_score
    results_dict_pca[k]['score'] = kmeans.score
    results_dict_pca[k]['model'] = kmeans
```

```python
for k in range(2, num_of_clusters):
    print('----------------------------------------------')
    print("Number of Clusters: {}".format(k))
    print('Silhouette Score (NOT PCA):', results_dict[k]['silhouette_score'])
    print('Silhouette Score (PCA):', results_dict_pca[k]['silhouette_score'])
```

```python
def encodeTimestamp(timestamp):
  # Take year, month, day, hour, minute, second from string
  # Intput: 2023-04-19 02:15:30+00:00
  #   year = 2023
  #   month = 4
  #   day = 19
  #   hour = 2
  #   minmute = 15
  #   second = 30
  Date = str(timestamp)

  year_month_day, hour_minute_second_timezone = str(Date).split(' ')
  hour_minute_second, timezone = hour_minute_second_timezone.split('+')

  year, month, day = [int(value) for value in year_month_day.split('-')]
  hour, minute, second = [int(value) for value in hour_minute_second.split(':')]

  # print(year, month, day, hour, minute, second)
  # Skip year
  # return year*12*30*24*60*60 + month*30*24*60*60 + day*24*60*60 + hour*60*60 + minute*60 + second

  return month*30*24*60*60 + day*24*60*60 + hour*60*60 + minute*60 + second
```

```python
lr_df = pd.DataFrame([],columns=temp_df.columns)

for row in range(len(df_price)):
  # 2023-04-19 02:15:30+00:00
  Date = temp_df.loc[row][0]

  process_Date = encodeTimestamp(Date)

  # Insert row to lr_df
  processed_row = [value for value in temp_df.loc[row][1:]]
  processed_row.insert(0, process_Date)

  lr_df.loc[len(lr_df)] = processed_row
```
